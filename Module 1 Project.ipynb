## Methodology

In order to answer these questions, we will follow the OSEMiN process which involves obtaining, scrubbing, exploring modeling, and interpreting our data.

OSEMN Process

Obtain - Make sure to understand the problem you are trying to solve first. Then, gather the information and data necessary to solve it. 

Scrub - Prepare your data so it is ready for processing. For example, identify and remove null values or outliers, normalize data, and feature engineer/feature select.

Explore - Understand the dataset you're working with by creating visualizations that quickly convey patterns in your data. For example, illustrate your distribution, heteroskedasticity, or multicollinearity. 

Model - Build your model and and tune it to best fit your data and provide the most accurate predictions. Return to the the scrub & expore stages, and iterate to optimize the input used in your modeling stage to result in better predictions.

Interpret - Digest the results of your model, and communicate your interpretation in a consumable way to stakeholders. When you have reached consensus between your stakeholders and decision scientists, you can proceed to productionize your model and automate the processes necessary to support it.

### Investigate Multicollinearity

Multicollinearity occurs when independent variables in a regression model are correlated. When variables are not independent, their correlation can cause the following problems: 
- Coefficients can swing dramatically based on other independent variables, making them sensitive to small changes in the model
- Multicollinearity reduces the precision of the estimate coefficients, weakening the statistical power of your model
- You might not be able to trust p-values and know which variables are actually statistically significant

The heatmap below underscores variables with correlationships greater than .75 with darker colors.

```python
# Heatmap to check for correlations
f, ax = plt.subplots(figsize=(7,7))
sns.heatmap(house_data.corr(), center=0, linewidths=.8, cmap="RdBu_r")
```

```python
# Isolate variables with a correlation greater than .75
f, ax = plt.subplots(figsize=(7,7))
sns.heatmap(abs(house_data.corr())>0.75, center=0, linewidths=.8, cmap="RdBu_r")
```

```python
# Display the correlation numbers for the above heatmap
house_data.iloc[:, :14].corr()
```

The variables with the greatest correlationship appear to be:

#### Predictability Statistics

```python
# Create Line of Best Fit

def linear_regression(X, y):
    # Split data set between training (70%) and test (30%)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
    linreg = LinearRegression()
    # Fit regression using training X, y data 
    linreg.fit(X_train, y_train)
    # Regression predicts y for each X in test set, which is compared to real y
    y_pred = linreg.predict(X_test)
    
# See split
print('Split')
print(X_test.shape)
print(X_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
# Measure performance of model
print('Predictability Statistics:')
print('R^2 Score:', round(linreg.score(X, y),2))
print('Mean Squared Error:', round((metrics.mean_squared_error(y_test, y_pred)),2))
print('Root Mean Squared Error:', round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),2))
print('Mean Absolute Error:', round(metrics.mean_absolute_error(y_test, y_pred),2))
print('Average Predicted Price:', round(y_pred.mean(),2))
print('Average Actual Price:', round(y_test.mean(), 2))
print('% Change in Y:', round(1 - (y_test.mean()/y_pred.mean()),4))
```

## Interpret Results

 ### Interpret OLS Model

Model Fit: Use MSE, RMSE or MAE

- MSE: MSE is a measure of how close a fitted line is to data points. If we unexpectedly have very large or small MSE, this can be a red flag that our regression line either fits the data too loosely (our erros are huge), or too tightly (we have an over-fit model that cannot generalize enough for future data). MSE also squares errors and is thusly skewed towards amplifying bad performing data points.

-> Our MSE is 52,442,762,386.51

- Root Mean Squared Error (RMSE): RMSE is the squared root of MSE and hence, returns the scale of the errors to the scale of targets, avoiding the overestimation of error tendency of MSE. We want to balance between overfiting (very low MSE for training data) and underfitting (very high MSE for test/validation data), so we will compare residuals for both test set and training set to ensure our model is optimally fit to our data. 

-> Our RME is 229,003.85

- Mean Absolute Error (MAE): MAE is average of the absolute differences between the target values and the predictions. It does not penalize huge erros and is not as sensitive to outliers like MSE is.

-> Our MAE is 149,393.98

Model Performance: Use r-Squared or adjusted r-squared

- R-squared: R^2 is the ratio between how good our model is versus the naive mean model. It ranges between negative infinity to 1, making it highly interpretable. Negative values denote our model is worse than the naive average model. As you add more predictors (independent variables) to the regression model, R-squared increases, falsely indicating predictability. This is why we need Adjusted R-squared.

-> Our R^2 is 61%

- Adjusted R-Squared: Adjusted R-squared compensates for the addition of variables and only increases if the new predictor enhances the model above what would be obtained by probability. Values will range from 0.0 to 1.0 for either R-squared or adjusted. For example, the adj r-squared of .61 below means the independent variables in our data explain 61% of the variation in price. 

-> Our Adjusted R^2 is also 61%

Model Significance: Use F-statistic and p-Value

- F-statistic: indicates whether your linear regression model provides a better fit to the data than a model that contains no independent variables. 

-> Our F-stat is 2359

- P-value: is the probability of observing a statistic at least as extreme as ours, conditional on the null hypothesis. A p-value doesn’t *prove* anything. It’s simply a way to use surprise as a basis for making a reasonable decision.
All of our values are < 0.05, denoting significance of all of our variables. 

Model Bias: Use Jarque-Bera (JB) and Durbin-Watson (DB)

- Jarque-Bera: is a goodness-of-fit test indicating whether sample data have the skewness and kurtosis matching a normal distribution. The further the value is from 0, the less normal the distribution is. 

-> Our JB is 404708.649

- Durbin Watson: is a test for autocorrelation in the residuals from a statistical regression analysis. The Durbin-Watson statistic will always have a value between 0 and 4, with a 2.0 meaning that there is no autocorrelation detected in the sample.

-> Our DW is 1.995
