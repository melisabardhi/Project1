## Methodology

In order to answer these questions, we will follow the OSEMiN process which involves obtaining, scrubbing, exploring modeling, and interpreting our data.

OSEMN Process

Obtain - Make sure to understand the problem you are trying to solve first. Then, gather the information and data necessary to solve it. 

Scrub - Prepare your data so it is ready for processing. For example, identify and remove null values or outliers, normalize data, and feature engineer/feature select.

Explore - Understand the dataset you're working with by creating visualizations that quickly convey patterns in your data. For example, illustrate your distribution, heteroskedasticity, or multicollinearity. 

Model - Build your model and and tune it to best fit your data and provide the most accurate predictions. Return to the the scrub & expore stages, and iterate to optimize the input used in your modeling stage to result in better predictions.

Interpret - Digest the results of your model, and communicate your interpretation in a consumable way to stakeholders. When you have reached consensus between your stakeholders and decision scientists, you can proceed to productionize your model and automate the processes necessary to support it.

Expensive (yellow/green) homes are clustered together in the same zipcodes. However, we can't distinguish the surrounding areas well. 
One option would be to use one-hot encoding, and convert the zipcode data into bins so that we can generate a linear model. However, doing this using lat and long will bin the data rectangularly, leading to model bias.
Another option would be to use spectral clustering to capture geographical clusters in our data. 
This will help us retain predictive power.

##### Spectral Clustering

```python
# Each unit x on the map is defined by the lat and long coordinates
X = house_data[['lat', 'long']]
```

```python
# Create 10 clusters
from sklearn.cluster import SpectralClustering
model = SpectralClustering(n_clusters=10, affinity='nearest_neighbors', assign_labels='kmeans')
labels = model.fit_predict(X)
```

```python
# Illustrate
plt.figure(figsize=(7, 7))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, s=10, cmap='RdYlGn')
plt.xlabel('long')
plt.ylabel('lat')
plt.title('10 Clusters of locations with varying price ranges')
```

Using these new 10 new cluster labels, we will create new columns to represent geographical location based on our cluster bins. This will allow us to build a linear relationship between price and location.

### Investigate Multicollinearity

Multicollinearity occurs when independent variables in a regression model are correlated. When variables are not independent, their correlation can cause the following problems: 
- Coefficients can swing dramatically based on other independent variables, making them sensitive to small changes in the model
- Multicollinearity reduces the precision of the estimate coefficients, weakening the statistical power of your model
- You might not be able to trust p-values and know which variables are actually statistically significant

### Feature Scaling, Normalization and Standardization

Feature Scaling: In scaling (also called min-max scaling), you transform the data such that the features are within a specific range, i.e., [0, 1].Feature scaling is another data pre-proceessing step applied to independent variable which helps move the data within a particular range and speeds up calculations in an algorithm.

Normalization: Changes your observations so that they can be described as a normal/Gaussian distribution, in which roughly equal observations fall above and below the means and your distribution is scaled with values between 0 and 1. 

Standardization (or z-score normalization): Transforms your data such that the resulting distribution has a mean of 0 and a standard deviation of 1.

```python
# Standardization of log transformed price, sqftlot, sqft_living
house_data['price'] = (price - np.mean(price)) / np.sqrt(np.var(price))
house_data['sqft_living'] = (sqft_living - np.mean(sqft_living)) / np.sqrt(np.var(sqft_living))
house_data['sqft_lot'] = (sqft_lot - np.mean(sqft_lot)) / np.sqrt(np.var(sqft_lot))
```

### Linear Regression Model

We have to split our data into training and test sets. We will use the training data to derive the model and test data to evaluate its goodness of fit.

```python
# Import libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import metrics
import statsmodels.api as sm
```

```python
# Assign variables to model
X = house_data.drop(['price'], axis=1)
y = house_data.price
```

```python
# Perform linear regression and validate results
#
def linear_regression(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    y_pred = linreg.predict(X_test)
    
    # See 75/25 split shape
    print('Split')
    print(X_test.shape)
    print(y_train.shape)
    print(y_test.shape)
    print(y_train.shape)
    print('-----------------------------------')

    # Predictability statistics
    print('Predictability Statistics')
    print('R^2:',round(linreg.score(X, y),2))
    print('Mean Absolute Error:', round(metrics.mean_absolute_error(y_test, y_pred),2))
    print('Root Mean Squared Error:', round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),2))
    print('Average Predicted Price:', round(y_pred.mean(),2))
    print('Average Actual Price:', round(y_test.mean(),2))
    
    # Display 100 house prices
    plt.figure(figsize=(16,6))
    sns.lineplot(range(len(y_pred[:200])), y_pred[:200], label='Predicted Prices')
    sns.lineplot(range(len(y_pred[:200])), y_test[:200], label='Actual Prices')
    plt.title('Comparing predicted price vs actual price', fontdict={'fontsize':20}), plt.xlabel('Values'), plt.ylabel('Prices')
    plt.legend(), plt.show()
    return (linreg, y_pred, y_test)
```

### Statistics for Evaluation 
Model Fit: Use RMSE
- Root Mean Squared Error (RMSE): summarizes all the residuals in the test and training set by telling you how close a regression line is to a set of points by taking the distances from the points to the regression line (or “errors”) and squaring them. We want to balance between overfiting (very low MSE for training data) and underfitting (very high MSE for test/validation data), so we will compare residuals for both test set and training set to ensure our model is optimally fit to our data. 

Model Performance: Use r-Squared or adjusted r-squared
- R-squared: the proportion of variance in the dependent variable (price) that is predictable from the independent variable (say, bedrooms). As you add more predictors (independent variables) to the regression model, R-squared increases, falsely indicating predictability. This is why we need Adjusted R-squared.
- Adjusted R-Squared: The adjusted R-squared compensates for the addition of variables and only increases if the new predictor enhances the model above what would be obtained by probability. Values will range from 0.0 to 1.0 for either R-squared or adjusted. For example, the adj r-squared of .60 below means the independent variables explain 60% of the variation in price. 

Model Significance: Use F-statistic and p-Value
- F-statistic: indicates whether your linear regression model provides a better fit to the data than a model that contains no independent variables. 
- P-value: is the probability of observing a statistic at least as extreme as ours, conditional on the null hypothesis. A p-value doesn’t *prove* anything. It’s simply a way to use surprise as a basis for making a reasonable decision.

Model Bias: Use Jarque-Bera (JB) and Durbin-Watson (DB)
- Jarque-Bera: is a goodness-of-fit test indicating whether sample data have the skewness and kurtosis matching a normal distribution. The further the value is from 0, the less normal the distribution is. 
- Durbin Watson: is a test for autocorrelation in the residuals from a statistical regression analysis. The Durbin-Watson statistic will always have a value between 0 and 4, with a 2.0 meaning that there is no autocorrelation detected in the sample.
